```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['age'] = df['age'].fillna (df['age'].median())
df['age'] = df['age'].astype(int)
 
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['gender'] = df['gender'].fillna (df['gender'].mode()[0])
df['country'] = df['country'].fillna (df['country'].mode()[0])
df['organ_type'] = df['organ_type'].fillna (df['organ_type'].mode()[0])
df['blood_type'] = df['blood_type'].fillna (df['blood_type'].mode()[0])
df['insurance_type'] = df['insurance_type'].fillna (df['insurance_type'].mode()[0])
df['pre_existing_conditions'] = df['pre_existing_conditions'].fillna (df['pre_existing_conditions'].mode()[0])
 
df['bmi'] = df['bmi'].fillna (df['bmi'].median())
df['match_score'] = df['match_score'].fillna (df['match_score'].median())
```
